```python
import FOCuS
import matplotlib.pyplot as plt
import numpy as np
```

```python
threshold=25
X = np.concatenate((np.random.normal(size=50), np.random.normal(size=50)+1.5))
changepoint, stopping_time, S_max = FOCuS.scan(X, threshold=threshold, plot=True, pre_change_mean=None)

plt.plot(X)
plt.title("Detected changepoint")
if changepoint is not None:
    plt.axvline(changepoint, color='red')
    plt.axvline(stopping_time, color='C1')
    plt.legend(['signal','changepoint', 'stopping time'])
```

```python
plt.plot(S_max)
plt.axhline(threshold, color="C1")
plt.title("Max $S_t$")
plt.xlabel("$t$")
```
